## Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

### Descriptive Analytics for Numerical Columns
+ Objective: To compute and analyze basic statistical measures for numerical columns in the dataset

```python
# import necessary library and read csv file
import pandas as pd
import numpy as np
data = pd.read_csv('sales_data_with_discounts.csv')
```

```python
data
```

```python
# Check data types for all columns
print(data.info())
```

```python
# Identify numerical columns in the dataset. 
numerical_columns = data.select_dtypes(include=[np.number]).columns
print("Numerical Columns:", numerical_columns)
```

```python
# Calculating mean for numerical columns.
mean_values = data[numerical_columns].mean()
print("Mean values for the numerical columns:")
print(mean_values)
```

```python
# Calculating median for numerical columns.
median_values = data[numerical_columns].median()
print("Median values for the numerical columns:")
print(median_values)
```

```python
# Calculating mode for numerical columns.
mode_values = data[numerical_columns].mode()
print("Mode values for the numerical columns:")
print(mode_values)
```

```python
# Calculating standard deviation for numerical columns.
standard_deviation = data[numerical_columns].std()
print("Standard Deviation values for the numerical columns:")
print(standard_deviation)
```

#### Brief interpretation of these statistics.
1. Volume: On average, 5 units are sold per order, but most orders are for 3 units. It varies a lot.  
2. Avg Price: The average price is ₹10,453, but most items are priced around ₹1,450. Prices differ greatly.  
3. Total Sales Value: Sales average ₹33,812 per transaction, but smaller sales around ₹5,700 are more common. There’s a huge variation.  
4. Discount Rate (%): Discounts average 15%, with 5% being the most common. Discounts are fairly consistent.  
5. Discount Amount: The average discount is ₹3,346, but many are smaller, around ₹989. Discount amounts vary widely.  
6. Net Sales Value: After discounts, the average transaction is ₹30,466, but smaller sales around ₹4,678 dominate. There’s a big range in values.  
+ conclusion: Sales and prices vary widely, with discounts influencing net values significantly.

### Data Visualization
+ Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.

#### Histograms: 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    data[column].plot(kind='hist', bins=20, edgecolor='black', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Calculating skewness for each numerical column
skewness = data[numerical_columns].skew()
print("Skewness of each numerical column:")
print(skewness)
```

```python
# Calculating outliers 
Q1 = data[numerical_columns].quantile(0.25)
Q3 = data[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
outliers_lower = Q1 - 1.5 * IQR
outliers_upper = Q3 + 1.5 * IQR

# Finding outliers for each column and printed
for column in numerical_columns:
    outlier_data = data[(data[column] < outliers_lower[column]) | (data[column] > outliers_upper[column])]
    print(f"Outliers for {column}:")
    print(outlier_data[column])  # Print the outlier values for the column
    print("\n")
```

#### Boxplots:

```python
import seaborn as sns

plt.figure(figsize=(15, 10))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=data[column], color='lightblue')
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```

+ Volume, Total Sales, and Discount Amount have multiple high outliers, indicating bulk purchases or special transactions.
+ Avg Price has extreme values above 50,000, suggesting premium-priced products.
+ Net Sales Value and Total Sales Value are right-skewed, meaning a few high-value transactions drive the sales.
+ Discount Rate (%) is well-distributed, but some low-discount outliers exist.


#### Barplots:

### Standardization of Numerical Variables
●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

### z-score normalization.
z-score normalization helps in following normal distribution
+ Befor standardization:These numbers are in percentage, small compared to sales prices.
+ After standardization: The numbers will be transformed to something like [-3,+3] Now, they’re on the same scale and easy to compare!

```python
# Standardize the numerical columns using the formula: z=x-mu/sigma
means = data[numerical_columns].mean()
stds = data[numerical_columns].std()

# Standardization (Z-score normalization)
standardized_data = data.copy()
standardized_data[numerical_columns] = (data[numerical_columns] - means) / stds
print(standardized_data[numerical_columns] )
```

```python
# Before and after comparisons of the data distributions.
fig, axes = plt.subplots(len(numerical_columns), 2, figsize=(8, 4* len(numerical_columns)))

for i, col in enumerate(numerical_columns):
    sns.histplot(data[col], bins=30, kde=True, ax=axes[i, 0], color='blue')
    axes[i, 0].set_title(f"Before Standardization: {col}")
    
    sns.histplot(standardized_data[col], bins=30, kde=True, ax=axes[i, 1], color='red')
    axes[i, 1].set_title(f"After Standardization: {col}")

plt.tight_layout()
plt.show()
```

### Conversion of Categorical Data into Dummy Variables
●	Objective: To transform categorical variables into a format that can be provided to ML algorithms.


### Discuss the need for converting categorical data into dummy variables ?
Converting categorical data into dummy variables is called as one-hot encoding. Its needed because most machine learning models work with numerical data, not text or categories.

```python
# Finding categorical columns in the dataset
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:", categorical_columns)
```

```python
#Applying one-hot encoding
df_encoded = pd.get_dummies(data, columns=categorical_columns)
print("\nTransformed Dataset:\n")
print(df_encoded.head())
```

###### Importance of Data Preprocessing Steps in Data Analysis and Machine Learning**  
Data preprocessing involves steps like Standardization , One-Hot Encoding and Handling Missing Values. Standardization for ensuring that numerical features are on a consistent scale, preventing models from being biased towards features with larger magnitudes. This is particularly important for algorithms like k-means clustering and gradient-based models. One-Hot Encoding will converts categorical variables into numerical form, allowing machine learning models to process them correctly. It prevents models from interpreting categorical values as ordinal and ensures better learning.
